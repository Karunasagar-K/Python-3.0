```python
import matplotlib.pyplot as pyplot
labels = ('python','Java','Scale','CB')
sizes = [45,30,15,10]

#only explode the 1st slice (i.e 'python')
explode = (0.1,0,0,0)
pyplot.pie(sizes,
           explode = explode,
          labels = labels,
          autopct ='%1.f%%',
          shadow = True,
          counterclock = False,
          startangle = 45)
pyplot.show()
```
